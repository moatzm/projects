```python
# Load data from the csv file
# Load packages
import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid", {"axes.facecolor": ".9"})

df = pd.read_csv(r"C:\Users\moatz\Desktop\New folder (5)\workspace (10)\workspace\House_price.csv")
df
```

```python
df.drop(columns='Unnamed: 6',inplace=True)
```

```python
import pandas as pd
import numpy as np
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.preprocessing import StandardScaler
```

```python
df['log'] = np.log(df['Price'])
```

```python
features = df.drop(columns=['Price','log'], axis=1)  # All columns except 'PRICE' will be features
target = df['log']  # 'PRICE' column is the target
scaler = StandardScaler()
# Fit the inputs (calculate the mean and standard deviation feature-wise)
scaler.fit(features)
inputs_scaled = scaler.transform(features)
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score


X_train, X_test, y_train, y_test = train_test_split(inputs_scaled, target, test_size=0.2, random_state=42)

svr = SVR()

# Initialize and train the SVR model
svr = SVR(kernel='rbf', C=1, epsilon=0.01)
svr.fit(X_train, y_train)

# Predict the target values for the test set
y_pred = svr.predict(X_test)

# Visualize the results
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, color='blue', edgecolor='k', alpha=0.7)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('SVR: Actual vs Predicted')
plt.show()
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R2 Score after tuning: {r2}')
```
